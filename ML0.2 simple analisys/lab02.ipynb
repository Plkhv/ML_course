11. Пусть задания с кратким ответом будут задания типа В. Соответственно всего по экзамену вопросов класса В: $В_1$, ... $В_к$ Посчитайте процент выполненных и невыполненных заданий по каждому вопросу класса В. 

```python
columns = [f"Задание {i}" for i in range(1, 26)] 
index = ["выполнило задание", "не выполнило задание"] 
df_rus = pd.DataFrame(0, index=index, columns=columns)

for i in range(short_answer_tasks_rus):
	completed = rus[(rus['Задания с кратким ответом'].str[i] != '-') & (rus['Задания с кратким ответом'].str[i] != '0')].shape[0]
	uncompleted = rus[(rus['Задания с кратким ответом'].str[i] == '-') | (rus['Задания с кратким ответом'].str[i] == '0')].shape[0]
	df_rus.loc["выполнило задание", f"Задание {i+1}"] = f'{completed/3048*100:.2f}%'
	df_rus.loc["не выполнило задание", f"Задание {i+1}"] = f'{uncompleted/3048*100:.2f}%'
print(df_rus)
```
