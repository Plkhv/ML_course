2. Создайте функцию, которая принимает как аргументы целое число N и первый элемент (вещественное число), и разность (вещественное число) и создает матрицу numpy по диагонали, которой располагаются первые N членов арифметической прогрессии.

```python
def create_diagonal_matrix(N, first_element, difference):

    progression = np.arange(N) * difference + first_element
    diagonal_matrix = np.diag(progression)
    
    return diagonal_matrix

# Пример использования
N = 5
first_element = 2.5
difference = 1.5
result_matrix = create_diagonal_matrix(N, first_element, difference)
print(f"Количество членов арифметической прогрессии: {N}")
print(f"Первый элемент: {first_element}")
print(f"Разница: {difference}")
print("Результирующая матрица:")
print(result_matrix)
```

5. Создайте две матрицы размером (5,5). Одна матрица содержит 5 в шахматном порядке как в задаче домашнего задания, другая имеет треугольную форму содержащую 5 на основной диагонали и в позициях выше ее, а ниже все 0. Посчитайте их детерминант и найдите обратные матрицы.

```python
chess_matrix = np.zeros((5, 5), dtype=int)
chess_matrix[::2, ::2] = 5  # четные строки и столбцы
chess_matrix[1::2, 1::2] = 5  # нечетные строки и столбцы

print("Матрица в шахматном порядке:")
print(chess_matrix)
```

```python
triangular_matrix = np.triu(np.full((5, 5), 5), k=0)

print("\nТреугольная матрица:")
print(triangular_matrix)
```

```python
det_chess = np.linalg.det(chess_matrix)

# Детерминант треугольной матрицы
det_triangular = np.linalg.det(triangular_matrix)

print("\nДетерминант шахматной матрицы:", det_chess)
print("Детерминант треугольной матрицы:", det_triangular)
```

```python
# Обратная матрица для треугольной матрицы
if det_triangular != 0:
    inverse_triangular = np.linalg.inv(triangular_matrix)
    print("\nОбратная треугольная матрица:")
    print(inverse_triangular)
else:
    print("\nОбратная матрица для треугольной матрицы не существует.")

# Обратная матрица для шахматной матрицы
if det_chess != 0:
    inverse_chess = np.linalg.inv(chess_matrix)
    print("\nОбратная шахматная матрица:")
    print(inverse_chess)
else:
    print("\nОбратная матрица для шахматной матрицы не существует.")
```

17. Постройте круговую диаграмму для признака Rooms, иллюстрирующую количество квартир в процентах в зависимости от количества комнат. Сделайте сектор с наибольшим числом квартир выдвинутым.

```python
import matplotlib.pyplot as plt
data_ = pd.read_csv(file_path)
# Группировка по количеству комнат и подсчет количества квартир
room_counts = data_['Rooms'].value_counts().sort_index()

# Определение сектора для выдвижения (с наибольшим числом квартир)
explode = [0] * len(room_counts)  # Создаем список с нулями

# Построение круговой диаграммы
plt.figure(figsize=(8, 8))
plt.pie(
    room_counts,
    labels=room_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    explode=explode,
    shadow=True
)
plt.title('Распределение квартир по количеству комнат', fontsize=16)

plt.show()
```

21. Постройте ящиковую диаграмму признака Square. Оцените визуально имеются ли выбросы, и, если да, то начиная с какого размера площади значение признака можно считать выбросом.

```python
# Настройка стиля графиков
sns.set(style="whitegrid")

# Построение ящиковой диаграммы
plt.figure(figsize=(13, 6))
sns.boxplot(x=data_['Square'], color='lightblue')
plt.title('Ящиковая диаграмма для признака Square', fontsize=16)
plt.xlabel('Square', fontsize=12)
plt.show()

# Вычисление границ выбросов
Q1 = data_['Square'].quantile(0.25)  # Первый квартиль
Q3 = data_['Square'].quantile(0.75)  # Третий квартиль
IQR = Q3 - Q1  # Межквартильный размах
lower_bound = Q1 - 1.5 * IQR  # Нижняя граница выбросов
upper_bound = Q3 + 1.5 * IQR  # Верхняя граница выбросов

print(f"Нижняя граница выбросов: {lower_bound}")
print(f"Верхняя граница выбросов: {upper_bound}")
```
